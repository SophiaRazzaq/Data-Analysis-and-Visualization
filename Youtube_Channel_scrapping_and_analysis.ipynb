#Scrapping the required data

```python
import time
from datetime import datetime, timedelta
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

start_date = datetime(2019, 9, 10)
end_date = datetime(2023, 9, 10)

current_date = datetime.now()

num_scrolls = 18
for _ in range(num_scrolls):
    driver.execute_script('window.scrollTo(0, document.documentElement.scrollHeight);')
    time.sleep(3)

video_elements = driver.find_elements(By.XPATH, '//*[@id="content"]/ytd-rich-grid-media')

video_titles = []
views_counts = []
upload_dates = []
likes_counts = []
video_links = []

for video_element in video_elements:
    video_title_element = video_element.find_element(By.XPATH, './/a[@id="video-title-link"]')
    video_title = video_title_element.text
    video_titles.append(video_title)
    print(video_title)

    views_count_element = video_element.find_element(By.CLASS_NAME, 'inline-metadata-item')
    views_count = views_count_element.text
    views_counts.append(views_count)
    print(views_count)

    upload_date_element = video_element.find_element(By.XPATH, './/*[@id="metadata-line"]/span[2]')
    upload_date = upload_date_element.text
    print(upload_date)

    if 'ago' in upload_date:
        # Calculate the actual upload date based on the current date
        days_ago = int(upload_date.split()[0])
        upload_date = current_date - timedelta(days=days_ago)
    else:
        # Convert the upload date to a datetime object
        upload_date = datetime.strptime(upload_date, '%b %d, %Y')
    upload_dates.append(upload_date)

    # Extract the video link
    video_link = video_title_element.get_attribute("href")
    video_links.append(video_link)
    print(video_link)

driver.quit()


print(len(video_titles))
print(len(views_counts))
print(len(upload_dates))
print(len(likes_counts))
print(len(video_links))
```

```python
for index, row in df.iterrows():
    video_link = row['Video Link']
    url = video_link
    driver.get(url)
    videos = driver.find_elements(By.CLASS_NAME, "style-scope ytd-rich-item-renderer")

    time.sleep(3)
    buttons = driver.find_elements(By.TAG_NAME, "button")
    for button in buttons:
            aria_label = button.get_attribute("aria-label")
            if aria_label and aria_label.startswith("like this video"):
                likes_count = aria_label.split()[5]
                likes_counts.append(likes_count)
                print(likes_counts)
                break
    df.at[index, 'Likes'] = likes_count
driver.quit()
```

```python
def outlierss(dfcopy):
    Q1 = dfcopy['Views'].quantile(0.25)
    Q3 = dfcopy['Views'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = dfcopy[(dfcopy['Views'] < lower_bound) | (dfcopy['Views'] > upper_bound)]
    print(" Data Frame after removing Outliers in the Views_count:")
    print(outliers)
```
